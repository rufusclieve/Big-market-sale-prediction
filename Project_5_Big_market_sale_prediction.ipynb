```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#  as there are missing value in the dataset
#1. using imputation technique we can fill the missing value
 #mean--> average value
 #mode--> most repeated value
 #median--> middle value

big_mart_data['Item_Weight'].mean()
```

```python
big_mart_data['Item_Weight'].fillna(big_mart_data['Item_Weight'].mean(), inplace=True)
```

```python
big_mart_data.isnull().sum()
```

```python
# replacing "Outlet_Size" with mode
mode_of_outlet_size = big_mart_data.pivot_table(values ='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode_of_outlet_size)
# this tells us that,
# in Grocery Store most of the time Outlet_Size is small, and Supermarket Type1 mode is small, etcc...
```

```python
missing_value = big_mart_data['Outlet_Size'].isnull()
print(missing_value)
# True means there is a missing value, Opposite for False
```

```python
big_mart_data.loc[missing_value, 'Outlet_Size'] = big_mart_data.loc[missing_value, 'Outlet_Type'].apply(lambda x: mode_of_outlet_size[x])
```

```python
big_mart_data.isnull().sum()
```

# **DATA PREPROCESSING**

```python
big_mart_data.head(2)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
big_mart_data.replace({'Item_Fat_Content': {'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'}}, inplace=True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

## **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
# for Item_Identifier
big_mart_data['Item_Identifier'] = encoder.fit_transform(big_mart_data['Item_Identifier'])

# for Item_Fat_Content
big_mart_data['Item_Fat_Content'] = encoder.fit_transform(big_mart_data['Item_Fat_Content'])

# for Item_Type
big_mart_data['Item_Type'] = encoder.fit_transform(big_mart_data['Item_Type'])

# for Outlet_Size
big_mart_data['Outlet_Size'] = encoder.fit_transform(big_mart_data['Outlet_Size'])

# for Outlet_Location_Type
big_mart_data['Outlet_Location_Type'] = encoder.fit_transform(big_mart_data['Outlet_Location_Type'])

# for Outlet_Type
big_mart_data['Outlet_Type'] = encoder.fit_transform(big_mart_data['Outlet_Type'])

# for Outlet_Identifier
big_mart_data['Outlet_Identifier'] = encoder.fit_transform(big_mart_data['Outlet_Identifier'])
```

```python
big_mart_data.head()
```

## Features and Target

```python
x = big_mart_data.drop(columns='Item_Outlet_Sales', axis=1)
y = big_mart_data['Item_Outlet_Sales']
```

```python
print(x)
```

```python
print(y)
```

## **Splitting training and testing dataset**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

## **Machine Learning Model training**

```python
regression = XGBRegressor()
```

```python
regression.fit(x_train, y_train)
```

## **Evaluation**

```python
# predictin on training data
training_data_prediction = regression.predict(x_train)
```

```python
# R square error
r2_error = metrics.r2_score(y_train, training_data_prediction)
print('R square error : ', r2_error)
```

```python
# predictin on testing data
testing_data_prediction = regression.predict(x_test)
```

```python
# R square error
r2_error = metrics.r2_score(y_test, testing_data_prediction)
print('R square error : ', r2_error)
```
